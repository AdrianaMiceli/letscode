```python
# Começamos sempre importando nossas bibliotecas.
import numpy as np
import pandas as pd
```

```python
#criar tabela com todos os MQLs, com todas as informações disponíveis para cada um.
```

```python
#ler os arquivos do dataset
df_closed_deals = pd.read_csv('olist_closed_deals_dataset.csv', sep=',', parse_dates=['won_date'])
df_marketing = pd.read_csv('olist_marketing_qualified_leads_dataset.csv', sep=',', parse_dates=['first_contact_date'])

```

```python
df_completo = pd.merge(df_marketing, df_closed_deals , how='left', on='mql_id')
df_completo.head()
```

```python
#criar uma coluna com quais MQLs fecharam acordo e quais foram perdidos.
mask = df_completo['sdr_id'].notna()
```

```python
mask
```

```python
df_completo['is_closed'] = mask
```

```python
df_completo.head()
```

```python
#Qual foi a taxa de conversão total?
mask = df_completo['is_closed']==True
taxa = len(df_completo[mask])/df_completo.shape[0]
print(f'A taxa de conversao foi de {taxa*100:.2f}%')
```

```python
df_origem_completo = df_completo.groupby('origin').agg({'is_closed':'count'})
```

```python
df_por_origem = df_completo[mask].groupby('origin').agg({'is_closed':'count'})
```

```python
#Qual foi a taxa de conversão de cada origem?
df_por_origem/df_origem_completo*100
```

```python
#Qual foi a taxa de conversão de cada página inicial?
df_lp_completo = df_completo.groupby('landing_page_id').agg({'is_closed':'count'})
df_por_lp = df_completo[mask].groupby('landing_page_id').agg({'is_closed':'count'})
df_por_lp/df_lp_completo*100
```

```python
#Para cada SR, qual a receita média declarada dos leads?
df_completo.groupby('sr_id').agg({'declared_monthly_revenue': 'mean'})
```

```python
#Para cada SR, quantos de cada business_type a pessoa converteu?
#df_completo[mask].groupby(['sr_id', 'business_type'])['is_closed'].count()
df_completo[mask].groupby(['sr_id', 'business_type']).agg({'is_closed':'count'})
```

```python
#Para cada SR, quantos de cada lead_behaviour_profile a pessoa converteu?
df_completo[mask].groupby(['sr_id', 'lead_behaviour_profile']).agg({'is_closed':'count'})
```

```python
#Para cada SR, quantos de cada lead_type a pessoa converteu?
df_completo[mask].groupby(['sr_id', 'lead_type']).agg({'is_closed':'count'})
```
